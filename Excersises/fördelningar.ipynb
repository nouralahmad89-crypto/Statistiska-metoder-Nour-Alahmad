```python
# Importerar pandas fr tabeller (DataFrame)
import pandas as pd

# Importerar NumPy fr numeriska berkningar och slumpgenerering
import numpy as np
import matplotlib.pyplot as plt

# Skapar en DataFrame med flera olika sannolikhetsfrdelningar
df = pd.DataFrame(
    np.column_stack([
        # 100 vrden frn Gammafrdelning (=2, =1)
        np.random.gamma(shape=2, scale=1, size=100),

        # 100 vrden frn standardnormalfrdelning (=0, =1)
        np.random.standard_normal(size=100),

        # 100 vrden frn en "vanlig" normalfrdelning (=10, =2)
        np.random.normal(loc=10, scale=2, size=100),

        # 100 vrden frn en uniform frdelning p intervallet [0, 1)
        np.random.uniform(low=0, high=1, size=100),

        # 100 vrden frn binomialfrdelning (n=5, p=0.3)
        np.random.binomial(n=5, p=0.3, size=100),

        # 100 vrden frn negativ binomialfrdelning
        # (antal misslyckanden fre r=3:e framgngen, p=0.4)
        np.random.negative_binomial(n=3, p=0.4, size=100),

        # 100 vrden frn geometrisk frdelning
        # (antal frsk tills frsta framgngen, p=0.4)
        np.random.geometric(p=0.4, size=100)
    ]),
    columns=[
        "Gamma",
        "Standard Normal",
        "Normal(mu=10,sigma=2)",
        "Uniform(0,1)",
        "Binomial(n=5,p=0.3)",
        "Nega Binomial(r=3,p=0.4)",
        "Geometric(p=0.4)"
    ]
)

# Ritar histogram fr varje kolumn
df.hist(figsize=(12, 9))
df.hist(cumulative= True)
plt.tight_layout()
```

```python
import scipy.stats as stats
import matplotlib.pyplot as plt

fig, ax = plt.subplots(3, 1, figsize=(7, 10))

stats.probplot(df["Gamma"].to_numpy(), plot=ax[0])
ax[0].set_title("Gamma")

stats.probplot(df["Standard Normal"].to_numpy(), plot=ax[1])
ax[1].set_title("Standard Normal")

stats.probplot(df["Binomial(n=5,p=0.3)"].to_numpy(), plot=ax[2])
ax[2].set_title("Binomial(n=5,p=0.3)")

plt.tight_layout()

```

En probplot (normal probability plot) anvnds fr att jmfra data direkt med en normalfrdelning.
Om datapunkterna ligger nra en rt linje r datan ungefr normalfrdelad.
Avvikelser frn linjen visar hur datan skiljer sig frn normalfrdelningen.
Nr stickprovsstorleken eller parametrar kar kan ven icke-normala frdelningar nrma sig en normalfrdelning.
