```python
import pandas as pd
import numpy as np

# Skapar en DataFrame med tre kolumner av slumpdata
df = pd.DataFrame(
    np.column_stack([
        # 100 slumpvärden från en Gammafördelning (α=2, β=1)
        np.random.gamma(2, 1, size=100),

        # 100 slumpvärden från en standardnormalfördelning (μ=0, σ=1)
        np.random.standard_normal(size=100),

        # 100 slumpvärden från en binomialfördelning (n=5, p=0.3)
        np.random.binomial(5, 0.3, size=100)
    ]),
    
    # Namn på kolumnerna i DataFrame
    columns=["Gamma/Chi^2", "Standard Normal", "Binomial"]
)

# Ritar histogram för varje kolumn i DataFrame
df.hist()
```
